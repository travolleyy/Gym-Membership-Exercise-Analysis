```python
df = pd.read_csv("healthcare_dataset.csv");df.head(10)
```

```python
plt.figure(figsize=(8, 6))

sns.histplot(df['Age'], kde=True, color='lightblue')

plt.title('Distribution of Age')

plt.xlabel('Age')

plt.ylabel('Frequency')

plt.show()
```

```python
plt.figure(figsize=(6, 4))
colors = ['red', 'blue']
sns.countplot(x='Gender', data=df, palette="Set2")
plt.title('Gender Distribution')
plt.xlabel('Gender')
plt.ylabel('Count')
plt.show()
```

```python
plt.figure(figsize=(8, 6))
sns.countplot(x='Blood Type', data=df, palette="Set2")
plt.title('Blood Type Distribution')
plt.xlabel('Blood Type')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(x='Age', y='Billing Amount', data=df)
plt.title('Billing Amount vs. Age')
plt.xlabel('Age')
plt.ylabel('Billing Amount')
plt.show()
```
